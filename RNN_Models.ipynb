```python
# Group by 'user' and take the top 12000 rows for each user
top_1200_per_user = final_df.groupby('session_encoded').head(12000)

# If you need to reset the index
top_1200_per_user = top_1200_per_user.reset_index(drop=True)

# Now top_1200_per_user contains the top 500 rows of each user
```

```python
# Group by 'session_encoded' and filter out groups with fewer than 1000 rows
test_final_df = test_final_df.groupby('session_encoded').filter(lambda x: len(x) >= 3000)

# Group by 'user' and take the top 12000 rows for each user
test_final_df = test_final_df.groupby('session_encoded').head(3000)

# If you need to reset the index
test_final_df = test_final_df.reset_index(drop=True)
```

```python
# Function to filter out the last two sessions for each user
def last_two_sessions(group):
    last_two_sessions = group['session_encoded'].unique()[-2:]
    return group[group['session_encoded'].isin(last_two_sessions)]

# Applying the function to each group and concatenating the results
validation_data = pd.concat([last_two_sessions(group) for _, group in top_1200_per_user.groupby('user_encoded')])

# Now, validation_data contains the last two sessions of each user
# To get the training data, you can just drop these rows from the original dataframe
training_data = top_1200_per_user.drop(validation_data.index)

# Resetting index if needed
validation_data.reset_index(drop=True, inplace=True)
training_data.reset_index(drop=True, inplace=True)
```

```python
# Separate features and target variable
# y_train = training_data['user_encoded']
y_train = training_data.groupby('session_encoded')['user_encoded'].first()
x_train = training_data.drop(['user_encoded', 'session_encoded'], axis=1)

# y_validation = validation_data['user_encoded']
y_validation = validation_data.groupby('session_encoded')['user_encoded'].first()
x_validation= validation_data.drop(['user_encoded','session_encoded'], axis=1)

print(x_train.shape[0]," ",x_train.shape[1]," ",y_train.shape[0])
print(x_validation.shape[0]," ",x_validation.shape[1]," ",y_validation.shape[0])
```

```python
# Normalize features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(x_train)
X_validation_scaled = scaler.fit_transform(x_validation)

# Reshape input to be [samples, time steps, features]
X_train_scaled = X_train_scaled.reshape((45, 12000, X_train_scaled.shape[1]))
X_validation_scaled = X_validation_scaled.reshape((20, 12000, X_validation_scaled.shape[1]))

y_train_encoded = to_categorical(y_train, 10)
y_validation_encoded = to_categorical(y_validation, 10)

print("shape of the train data that is passed to the LSTM Model")
print(X_train_scaled.shape[0]," ",X_train_scaled.shape[1]," ",X_train_scaled.shape[2])
# print(X_validation_scaled.shape[0]," ",X_validation_scaled.shape[1]," ",y_validation_encoded.shape[0])

# Initialize the RNN
model = Sequential()

# Adding the first LSTM layer
model.add(LSTM(units=50, activation='tanh', return_sequences=False, input_shape=(X_train_scaled.shape[1], X_train_scaled.shape[2])))
# model.add(LSTM(units=20, return_sequences=False))

# Adding the output layer
model.add(Dense(10, activation='softmax'))

# Compiling the RNN
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Fitting the RNN to the Training set
# model.fit(X_train_scaled, y_train_encoded, epochs=200, batch_size=45)
model.fit(X_train_scaled, y_train_encoded, epochs=100, batch_size=num_unique_sessions_train, validation_data=(X_validation_scaled, y_validation_encoded))
```
